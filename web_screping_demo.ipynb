```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np

# get webpage
webpage=requests.get('http://books.toscrape.com/').text
soup=BeautifulSoup(webpage,'lxml')

# check connection
print(soup.title.text)
```

```python
# find all book containers
soup.find_all('article',class_='product_pod')
```

```python
# get all books
books=soup.find_all('article',class_='product_pod')

# get the first book title
print(books[0].find('h3').find('a')['title'])

# get the first book price
print(books[0].find('p',class_='price_color').text)

# get the first book rating
print(books[0].find('p',class_='star-rating')['class'][1])
```

```python
# empty lists
title=[]
price=[]
rating=[]

for i in books:
  try:
    title.append(i.find('h3').find('a')['title'])
  except:
    title.append(np.nan)

  try:
    price.append(i.find('p',class_='price_color').text)
  except:
    price.append(np.nan)

  try:
    rating.append(i.find('p',class_='star-rating')['class'][1])
  except:
    rating.append(np.nan)

# check how many data having
print(len(title))
print(len(price))
print(len(rating))
```

```python
# create dataframe
df=pd.DataFrame({'Title':title,
                 'Price':price,
                 'Rating':rating})

print(df.head(10))
```

```python
# check pagination number
pagination=soup.find('li',class_='current').text.strip()
print(pagination)
```

```python
final=pd.DataFrame()

for j in range(1,51):
  webpage=requests.get('http://books.toscrape.com/catalogue/page-{}.html'.format(j)).text
  soup=BeautifulSoup(webpage,'lxml')
  books=soup.find_all('article',class_='product_pod')

  title=[]
  price=[]
  rating=[]
  availability=[]

  for i in books:
    try:
      title.append(i.find('h3').find('a')['title'])
    except:
      title.append(np.nan)

    try:
      price.append(i.find('p',class_='price_color').text)
    except:
      price.append(np.nan)

    try:
      rating.append(i.find('p',class_='star-rating')['class'][1])
    except:
      rating.append(np.nan)

    try:
      availability.append(i.find('p',class_='instock availability').text.strip())
    except:
      availability.append(np.nan)

  df=pd.DataFrame({'Title':title,
                   'Price':price,
                   'Rating':rating,
                   'Availability':availability})

  final=pd.concat([final,df],ignore_index=True)

  print(f'Page {j} done')

print(f'\nTotal books scraped: {len(final)}')
print(final.head())
```

```python
# save to csv
final.to_csv('books_data.csv',index=False)

print('File saved successfully!')
print(f'Total rows: {len(final)}')
print(f'Columns: {list(final.columns)}')

# data info
print('\nData Info:')
print(final.info())
```
